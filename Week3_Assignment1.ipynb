Part B: Exploratory Data Analysis (EDA)


```python
!pip install plotly matplotlib seaborn --quiet
import plotly.express as px
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
#Q4. Distribution Analysis
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
#Create histograms for:

#rainfall_mm
sns.histplot(data=agriculture_df, x='rainfall_mm', kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribution of Rainfall (mm)')

# temperature_c
sns.histplot(data=agriculture_df, x='temperature_c', kde=True, ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title('Distribution of Temperature (°C)')

# fertilizer_kg
sns.histplot(data=agriculture_df, x='fertilizer_kg', kde=True, ax=axes[1, 0], color='orange')
axes[1, 0].set_title('Distribution of Fertilizer (kg)')

# yield_ton_per_hectare
sns.histplot(data=agriculture_df, x='yield_ton_per_hectare', kde=True, ax=axes[1, 1], color='salmon')
axes[1, 1].set_title('Distribution of Yield (Ton/Hectare)')

plt.tight_layout()
plt.show()
```

OBSERVATIONS

1. rainfall_mm
*   The rainfall values are spread evenly from 300 mm to 1200 mm, showing a wide range of values.
*   There is no peak or skewness in the distribution showing a balanced distribution.

 2. temperature_c


*   The temperature values are distributed from about 18°C to 38°C, showing all cool to warm agricultural conditions.
*   The graph is nearly uniform with all the temperature ranges.

3. fertilizer_kg


*   Fertilizer usage ranges from about 50 kg to 250 kg.
*   This distribution is also even


4.  yield_per_ton_hectare


*   This distribution is approximately bell-shaped (normal) and with a mean of 5 tons/hectare, showing that most farms achieve yields near this value.
*   Very few observations are below 3 tons/hec or above 7 tons/hec, showing that very high or low yield values are rare.
*  there are no major outliers visible in the histogram.

```python
#Q9. Correlation Analysis

# a. Generate a correlation matrix for numerical features.
corr_matrix = agriculture_df.select_dtypes(include=['int64', 'float64']).corr()

# b.Create a heatmap.
sns.heatmap(corr_matrix, cmap='Blues', annot=True)
plt.title('Correlation Matrix');

```

```python
# c.Identify the top three features most correlated with crop yield.
# Correlation with crop yield
yield_corr = corr_matrix['yield_ton_per_hectare'].drop('yield_ton_per_hectare')

# Top 3 correlated features
top_3 = yield_corr.abs().sort_values(ascending=False).head(3)

print("Top 3 Features Most Correlated with Yield:")
print(top_3)
```

```python
#Q10. Group-Based Analysis
#Calculate the average yield for:
#a. Each crop type
crop_yield = agriculture_df.groupby('crop_type')['yield_ton_per_hectare'].mean()

print("Average Yield by Crop Type:")
print(crop_yield)

#b. Each soil type
soil_yield = agriculture_df.groupby('soil_type')['yield_ton_per_hectare'].mean()

print("\nAverage Yield by Soil Type:")
print(soil_yield)

```

```python
#check for missing values
agriculture_df.isnull().sum()
```

```python
#Q11. Feature Encoding

# Identify the categorical columns.
cat_cols = agriculture_df.select_dtypes(include='object').columns
print(cat_cols)

# Convert them into numerical form using One-Hot Encoding.
agri_encoded = pd.get_dummies(
    agriculture_df,
    columns=['crop_type', 'soil_type'],
    drop_first=True
)
# Display the first five rows of the transformed dataset.
print(agri_encoded.head())
```

Q12. Feature Selection

Separate:

 Input features (X)

 Target variable (y)

Specify which column is being used as the target variable.

```python
# Target variable
y = agri_encoded['yield_ton_per_hectare']

# Input features
X = agri_encoded.drop('yield_ton_per_hectare', axis=1)

print("Target Variable:")
print("yield_ton_per_hectare")

print("\nInput Features:")
print(X.columns.tolist())
```

Part D: Machine Learning

Q13. Train-Test Split

Split the dataset into:

80% Training Data

20% Testing Data

Display the shape of:

 X_train

 X_test

 y_train

 y_test


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
print("X_train Shape:", X_train.shape)
print("X_test Shape:", X_test.shape)
print("y_train Shape:", y_train.shape)
print("y_test Shape:", y_test.shape)
```

```python
#Q14. Linear Regression Model

# Train a Linear Regression model.
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()

lr_model.fit(X_train, y_train)

#Display the model coefficients and intercept.
print("Intercept:")
print(lr_model.intercept_)

print("\nCoefficients:")
print(lr_model.coef_)

#creating a coefficient table
coef_df = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': lr_model.coef_
})

coef_df = coef_df.sort_values(
    by='Coefficient',
    ascending=False
)

print(coef_df)

#Which feature has the highest positive coefficient?
highest_feature = coef_df.iloc[0]

print("\nFeature with Highest Positive Coefficient:")
print(highest_feature['Feature'])

print("\nCoefficient Value:")
print(highest_feature['Coefficient'])
```
